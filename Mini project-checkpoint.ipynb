```python
# Student Performance Data Analysis & Visualization
# This mini project demonstrates data science concepts (lists, arrays, dictionaries, NumPy, Pandas, broadcasting, file handling, visualization).
import pandas as pd

# Create dataset using dictionary
data = {
    "Name": ["Ritu", "Neha", "Neetu", "Annu"],
    "Class": ["MCA", "MCA", "MCA", "MCA"],
    "Marks": [90, 85, 88, 76],
    "Hobbies": ["Coding", "Music", "Coding", "Reading"]
}

df = pd.DataFrame(data)
print(df)
```

```python
import numpy as np
marks = np.array(df["Marks"])
avg_marks = np.mean(marks)
print("Average Marks:", avg_marks)
```

```python
high_scorers = df[df["Marks"] > 80]
print(high_scorers)
```

```python
bonus = np.array([5])   # scalar broadcasted
df["Final Marks"] = df["Marks"] + bonus
print(df)
```

```python
df.to_csv("student_data.csv", index=False)
```

```python
new_df = pd.read_csv("student_data.csv")
print(new_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar chart of marks
plt.bar(df["Name"], df["Marks"])
plt.title("Student Marks Comparison")
plt.xlabel("Name")
plt.ylabel("Marks")
plt.show()

# Distribution plot
sns.histplot(df["Marks"], kde=True)
plt.title("Marks Distribution")
plt.show()
```

```python
df["Maths"] = [85, 78, 92, 70]
df["Python"] = [90, 88, 95, 80]
df["DBMS"] = [88, 82, 89, 75]

print(df.mean(numeric_only=True))   # subject-wise average
```

```python
import numpy as np

# Average marks using NumPy
marks = np.array(df[["Maths","Python","DBMS"]])
avg_marks = np.mean(marks, axis=0)
print("Subject-wise Average:", avg_marks)

# Filter high scorers
high_scorers = df[df["Python"] > 85]
high_scorers
```

```python
print(df[["Maths","Python","DBMS"]].corr())
```

```python
# Save to CSV
df.to_csv("student_data.csv", index=False)

# Read back
new_df = pd.read_csv("student_data.csv")
new_df
```

```python
sns.heatmap(df[["Maths","Python","DBMS"]].corr(), annot=True)
plt.show()

df["Hobbies"].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```
